```python
import pandas as pd

# Create a DataFrame from the monthly_sales_data csv file
# Ask it to convert all of the dates in the file from strings to dates
df = pd.read_csv("monthly_sales_data.csv", parse_dates=['Date'], index_col=["Date"])
```

```python
# Create a histogram for one category
fig = df["Women's Clothing"].plot(kind='hist', bins=9, alpha=0.6, figsize=(15,12), 
                                  title="Sales for Womens Clothing")

# xlabel cannot be set when creating the plot. Use set_xlabel to set after
fig.set_xlabel("Monthly Sales Amount (USD)")

# Used the savefig method to save the chart to a file
fig.figure.savefig('hist_monthly.png')
```

```python
import pandas as pd

df = pd.read_csv("sales_data.csv", parse_dates=['Date'], index_col=['Date'])

# Process Data So it is Suitable for a line plot
# Use group by to get total sales by day
daily_sales = df.groupby('Date')['Sales'].sum()

# Create a histogram
fig = daily_sales.plot(kind='hist', bins=13, alpha=0.6, figsize=(15,12), 
                       title="Daily Sales", xlabel='Sales (USD)')

# xlabel cannot be set when creating the plot. Use set_xlabel to set after
fig.set_xlabel("Sales (USD)")

# Used the savefig method to save the chart to a file
fig.figure.savefig('hist_daily.png')
```
