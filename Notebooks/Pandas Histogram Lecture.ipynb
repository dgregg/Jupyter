      <li>Data File: <a href="data/monthly_sales_data.csv" target="_blank" rel="noopener">monthly_sales_data.csv</a></li>
    <li>Below we show the a simple histogram for "Women's Clothing"&nbsp; it examines if the monthly sales for "Women's Clothing" is normally distributed - that is if it is clustered around a central number.</li>
    <li>In the histogram below, you can see that the center is near 14,000. However, unlike the histogram above, the data is not grouped near the center number and there are more values to the right of 14000 than to the left.</li>
    <li>This is probably because the dataset only has 12 monthly values - histograms work better with more data.</li>

```python
import pandas as pd

# Create a DataFrame from the monthly_sales_data csv file
# Ask it to convert all of the dates in the file from strings to dates
df = pd.read_csv("data/monthly_sales_data.csv", parse_dates=['Date'], index_col=["Date"])
```

```python
# Create a histogram for one category
fig = df["Clothing"].plot(kind='hist', bins=9, alpha=0.6, figsize=(15,12), 
                                  title="Sales for Clothing & Shoes")

# xlabel cannot be set when creating the plot. Use set_xlabel to set after
fig.set_xlabel("Monthly Sales Amount (USD)")

# Used the savefig method to save the chart to a file
fig.figure.savefig('hist_monthly.png')
```

<h3>&nbsp;</h3>
<h3>Daily Sales</h3>
<ul>
    <li>Instead of looking at monthly sales, it is better to look at daily sales when using a histogram.</li>
    <li>The code below uses the pandas library to read a CSV file named "sales_data.csv" and process it for creating a histogram.</li>
    <li>First, the CSV file is read into a pandas DataFrame using the <code>pd.read_csv</code> method. The <code>parse_dates</code> parameter is set to <code>['Date']</code> to parse the <code>Date</code> column as a date, and the <code>index_col</code> parameter is set to <code>['Date']</code> to set the <code>Date</code> column as the index of the DataFrame.</li>
    <li>Next, the DataFrame is grouped by date and the sum of sales for each date is calculated using the <code>groupby</code> method and the <code>sum</code> function. This creates a new DataFrame named <code>daily_sales</code> that has a single column named 'Sales', indexed by date.</li>
    <li>Then, a histogram is created using the <code>plot</code> method of the <code>daily_sales</code> DataFrame. The parameters for the <code>plot</code> method are set as follows:</li>
</ul>
<ul>
    <li style="list-style-type: none;">
        <ul>
            <li><code>kind='hist'</code>: create a histogram plot</li>
            <li><code>bins=13</code>: use 13 bins for the histogram</li>
            <li><code>alpha=0.6</code>: set the transparency level to 0.6</li>
            <li><code>figsize=(15,12)</code>: set the size of the plot to 15x12 inches</li>
            <li><code>title="Daily Sales"</code>: set the title of the plot to "Daily Sales"</li>
            <li><code>xlabel='Sales (USD)'</code>: set the label of the x-axis to "Sales (USD)"</li>
        </ul>
    </li>
    <li>Finally, the <code>savefig</code> method of the plot object <code>fig</code> is used to save the histogram plot as a PNG file named "hist_daily.png".</li>
    <li>In the histogram below, you can see that the center is near 1,000.</li>
    <li>It is a right-skewed distributions, the long tail extends to the right while most values cluster on the left, as shown below. This suggests that while on average the store sells about $1000 a day, there are some peak days where the sales are 3 to 4 times that.</li>
</ul>

```python
import pandas as pd

df = pd.read_csv("data/sales_data.csv", parse_dates=['Date'], index_col=['Date'])

# Process Data So it is Suitable for a line plot
# Use group by to get total sales by day
daily_sales = df.groupby('Date')['Sales'].sum()

# Create a histogram
fig = daily_sales.plot(kind='hist', bins=13, alpha=0.6, figsize=(15,12), 
                       title="Daily Sales", xlabel='Sales (USD)')

# xlabel cannot be set when creating the plot. Use set_xlabel to set after
fig.set_xlabel("Sales (USD)")

# Used the savefig method to save the chart to a file
fig.figure.savefig('hist_daily.png')
```
