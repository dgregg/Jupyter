<h3 id="8045" class="on ki gm bf kj oo op hn km oq or hq kp hr os ht kt hu ot hw kx hx ou hz lb ov hj">Example 1: Loading JSON to Python list of dictionaries</h3>
<ul>
<li id="38ff" class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">Start by importing the <code class="nt nz oa ob oc b">json</code> library.</li>
<li class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">We use the function <code class="nt nz oa ob oc b">open</code> to open the JSON file and then the method <code class="nt nz oa ob oc b">json.load()</code> to parse the JSON string into a Python dictionary called js.</li>
<li class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">That&rsquo;s it! You now have a list of Python dictionaries from your JSON file.</li>
<li class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">One thing of note is that the <code class="nt nz oa ob oc b">json</code> library has both <code class="nt nz oa ob oc b">load()</code> and <code class="nt nz oa ob oc b">loads()</code>.</li>
<li class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">Both do the same thing, but <code class="nt nz oa ob oc b">loads()</code> is to create a Python object from a JSON string whereas <code class="nt nz oa ob oc b">load()</code> is to create a Python object from a JSON file.</li>
<li class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">You can think of the extra &lsquo;s&rsquo; in <code class="nt nz oa ob oc b">loads()</code> as &ldquo;load for strings&rdquo;.</li>
</ul>


```python
# import json module
import json

# Read Simple JSON Format File
jfile = open('data/Colorado_county_race.json')

# Load the JSON text from the file into a list of dictionaries using the json library
js = json.load(jfile)
jfile.close()

print(js[0].keys())

# Output data from 17th Record
# use index [16] followed by key for item in nested dictionary
print('\n\n' + js[16]['County'])
print('Total population', js[16]['Total population'])
print('Food Insecurity population', js[16]['Food Insecurity population'])
print('Food Insecurity Rate', js[16]['Food Insecurity Rate'])
print('Black poverty population', js[16]['Black poverty population'])
print('Black poverty rate', js[16]['Black poverty rate'])
print('White nonhispanic poverty population', js[16]['White nonhispanic poverty population'])
print('White nonhispanic poverty rate', js[16]['White nonhispanic poverty rate'])

# Loop through all records (break after first one)
for row in js :
    print('\n\n' + row['County'])
    print('Total population', row['Total population'])
    print('Food Insecurity population', row['Food Insecurity population'])
    print('Food Insecurity Rate', row['Food Insecurity Rate'])
    print('Black poverty population', row['Black poverty population'])
    print('Black poverty rate', row['Black poverty rate'])
    print('White nonhispanic poverty population', row['White nonhispanic poverty population'])
    print('White nonhispanic poverty rate', row['White nonhispanic poverty rate'])
    break
```


<ul>
<li>We can access data anywhere in our JSON file by chaining together the key names and/or indices.</li>
<li>For example, let&rsquo;s say we want to access the data for <span style="background-color: #eccafa;">Denver County</span> which is the 17th county in our file.
<ul>
<li>We can get ANY data for Denver by using the name of the json data list (js), followed by the index for Denver [16] followed by the "key" for the data we want to extract.</li>
<li><span style="background-color: #eccafa;">js[16]['County']</span> returns the county name <strong>Denver County</strong></li>
<li><span style="background-color: #eccafa;">js[16]['Total population']</span> returns the total population for Denver County <strong>705576</strong></li>
</ul>
</li>
<li><strong></strong>You can also loop through the list of data and print the results for each county
<ul>
<li>We can get ANY data for county by using the name of the loop variable (row), followed by the "key" for the data we want to extract.</li>
<li><span style="background-color: #bfedd2;">row['County']</span> returns the county name <strong>Adams County</strong> the fist time through the loop<strong><br /></strong></li>
<li><span style="background-color: #bfedd2;">row['Total population']</span> returns the total population for Adams County <strong>50410</strong>8 the fist time through the loop<strong></strong></li>
</ul>
</li>
</ul>

<li class="jl jm gm jn b hl nd jp jq ho ne js jt ju nf jw jx jy ng ka kb kc nh ke kf kg gf hj">We use the function <code class="nt nz oa ob oc b">open</code> to open the JSON file and then the method <code class="nt nz oa ob oc b">json.load()</code> to parse the JSON string into a Python dictionary called js.</li>
</ul>





```python
# import json module 
import json 

# Read Nested JSON Format
jfile = open('data/Colorado_county_race_grouped.json')

# Load the JSON text from the file into a list of dictionaries using the json library
js = json.load(jfile)
jfile.close()
        


print(js[0].keys())

# Output data from 17th Record
# use index [16] followed by key for item in nested dictionary
print('\n\n' + js[16]['County'])
print('Total population', js[16]['Total population'])
print('Food Insecurity population', js[16]['Food Insecurity population'])
print('Food Insecurity Rate', js[16]['Food Insecurity Rate'])

#  use index [16] followed by key for nested dictionaries 
# 'poverty population' and 'poverty rate' are nested dictionaries (use these keys)
# "'Black' and 'White nonhispanic' are keys in the 'poverty population' and 'poverty rate' dictionaries
print('poverty population', 'Black', js[16]['poverty population']['Black'])
print('poverty rate', 'Black', js[16]['poverty rate']['Black'])
print('poverty population', 'White nonhispanic', js[16]['poverty population']['White nonhispanic'])
print('poverty rate', 'White nonhispanic', js[16]['poverty rate']['White nonhispanic'])

# Loop through all records (break after first one)
for row in js :
    print('\n\n' + row['County'])
    print('Total population', row['Total population'])
    print('Food Insecurity population', row['Food Insecurity population'])
    print('Food Insecurity Rate', row['Food Insecurity Rate'])
    
    # 'poverty population' and 'poverty rate' are dictionaries containing data for all races
    pp = row['poverty population'] # Extract poverty population dictionary
    pr = row['poverty rate']       # Extract poverty rate dictionary
    print()
    print(pp.keys())
    
    # extract data from 'poverty population' and 'poverty rate' dictionaries
    print('\npoverty population', 'Black',  pp['Black'])
    print('poverty rate', 'Black', pr['Black'])
    print('poverty population', 'White nonhispanic', pp['White nonhispanic'])
    print('poverty rate', 'White nonhispanic',  pr['White nonhispanic'])
    break
```
