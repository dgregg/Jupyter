<h2>List Concatenation</h2>
<ul>
<li>We can create a new list by adding two existing lists together using the concatenation operator (+)</li>
</ul>
<pre>a = [1, 2, 3]<br />b = [4, 5, 6]<br />c = a + b<br />print(c)<br />print(a)</pre>
<p><strong>Output:</strong></p>
<pre><span style="color: #0000ff;">[1, 2, 3, 4, 5, 6]<br /></span><span style="color: #0000ff;">[1, 2, 3]</span></pre>
<hr />
<h2>List Slicing</h2>
<ul>
<li>Lists can be sliced using :</li>
<li>Remember:&nbsp; Just like in strings, the second number is &ldquo;up to but not including&rdquo;</li>
<li>If you leave one of the numbers off the slice goes to the beginning or the end of the list.</li>
</ul>
<pre>t = [9, 41, 12, 3, 74, 15]<br />print(t[1:3])  <span style="color: #999999;"><em>#[41,12]</em></span><br />print(t[:4]) <em><span style="color: #999999;">#[9, 41, 12, 3]</span></em><br />x= t[3:]<br />print(x) <em><span style="color: #999999;">#[3, 74, 15]</span></em><br />print(t[:]) <span style="color: #999999;"><em>#[9, 41, 12, 3, 74, 15]<br /></em></span></pre>
<p><strong>Output</strong></p>
<pre>[41, 12]<br />[9, 41, 12, 3]<br />[3, 74, 15]<br />[9, 41, 12, 3, 74, 15]</pre>
<hr />
<h2>in / not in</h2>
<ul>
<li>Python provides two operators that let you check if an item is in a list</li>
<li>These are logical operators that return True or False</li>
<li>They do not modify the list</li>
</ul>
<pre>nums = [1, 9, 21, 10, 16]<br />print(9 in nums)<br />print(15 in nums)<br />print(20 not in nums)</pre>
<p><strong>Output:</strong></p>
<pre><span style="color: #0000ff;">True<br /></span><span style="color: #0000ff;">False<br /></span><span style="color: #0000ff;">True</span> </pre>

<h2>Built-in List Functions</h2>
<ul>
<li>There are a number of functions built into Python that take lists as parameters
<ul>
<li>cmp(list1, list2): Compares elements of both lists.</li>
<li>len(list): Gives the total length of the list.</li>
<li>max(list): Returns item from the list with max value.</li>
<li>min(list): Returns item from the list with min value.</li>
<li>list(seq): Converts a tuple into list.</li>
</ul>
</li>
</ul>
<pre>nums = [3, 41, 12, 9, 74, 15]<br />print("len", len(nums))<br />print("max", max(nums))<br />print("min", min(nums))<br />print("sum", sum(nums))<br />print("avg", sum(nums)/len(nums))</pre>
<p><strong>Output:</strong></p>
<pre><span style="color: #0000ff;">len 6<br />max 74<br />min 3<br />sum 154<br />avg 25.666666666666668</span></pre>

<h2>List Methods</h2>
<ul>
<li><span>Just like string methods, list methods are working on the list it's being called from, if you for example have a list called: list1 = ["Movies", "Music", "Pictures"], then the list method is called such this: list1.list_method() </span></li>
</ul>
<p><span>List methods include:</span></p>
<ul>
<li>list.append(obj): Appends object obj to list</li>
<li>list.count(obj): Returns count of how many times obj occurs in list</li>
<li>list.extend(seq): Appends the contents of seq to list</li>
<li>list.index(obj): Returns the lowest index in list that obj appears</li>
<li>list.insert(index, obj): Inserts object obj into list at offset index</li>
<li>list.pop(obj=list[-1]): Removes and returns last object or obj from list</li>
<li>list.remove(obj): Removes object obj from list</li>
<li>list.reverse(): Reverses objects of list in place</li>
<li>list.sort([func]): Sorts objects of list, use compare func if given<br /><br /></li>
</ul>
<h3>append</h3>
<ul>
<li>We can create an empty list and then add elements using the append method</li>
<li>The list stays in order and new elements are added at the end of the list</li>
</ul>
<pre>stuff = list()<br />stuff.append('book')<br />stuff.append(99)<br />print(stuff)</pre>
<p><strong>Output</strong></p>
<pre><span style="color: #0000ff;">['book', 99]</span></pre>
<pre>stuff.append('cookie')<br />print(stuff)</pre>
<p><strong>Output</strong></p>
<pre><span style="color: #0000ff;">['book', 99, 'cookie']</span></pre>
<h3>sort</h3>
<ul>
<li>A list can hold many items and keeps those items in the order until we do something to change the order</li>
<li>A list can be sorted <br />(i.e., change its order)</li>
<li>The sort method (unlike in strings) means &ldquo;sort yourself&rdquo;</li>
</ul>
<pre>friends = [ 'Joseph', 'Glenn', 'Sally' ]<br />friends.sort()<br />print(friends)</pre>
<p><strong>Output:</strong></p>
<pre><span style="color: #0000ff;">['Glenn', 'Joseph', 'Sally']</span></pre>
<pre>print(friends[1])</pre>
<p><strong>Output:</strong></p>
<pre><span style="color: #0000ff;">Joseph</span></pre>
