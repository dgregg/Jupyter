```python
import pandas as pd

# Create a DataFrame from the monthly_sales_data csv file
# Ask it to convert all of the dates in the file from strings to dates
df = pd.read_csv("monthly_sales_data.csv", parse_dates=['Date'], index_col=["Date"])

display(df.head())
```

```python
# Create an area chart, assign the chart to a variable fig
fig = df.plot(kind='area', figsize=(12,15), ylabel='Sales (USD)', 
              title='Stacked Area chart: Monthly Sales Amount by Category Type')

# Used the savefig method to save the chart to a file
fig.figure.savefig('area_stacked.png')
```

```python
# Create a non-stacked area chart
fig = df.plot(kind='area', stacked=False, figsize=(15,12), ylabel='Sales (USD)', 
              title='Area chart: Monthly Sales Amount by Category Type')

# Used the savefig method to save the chart to a file
fig.figure.savefig('area_notstacked.png')
```
