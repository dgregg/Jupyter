```python
#import numpy as np
import pandas as pd

whisky = pd.read_csv("whiskies.txt", index_col=0)

# Add a second file as an extra column
whisky["region"] = pd.read_csv("regions.txt")

# Inspect the DataFrame using head and tail.
display(whisky.head())
display(whisky.tail())
```

```python
# iloc gets rows (or columns) at an index 
# it only takes integers (unlike loc).
display(whisky.iloc[0:5])
```

```python
# Specify row indices, column indices to get both
display(whisky.iloc[5:10, 0:5])
```

```python
# Index & Column names
display(whisky.index)
display(whisky.columns)
```

```python
# Create a named view of DataFrame
flavors = whisky.iloc[:, 2:14]
display(flavors.tail())
```

```python
# describe() shows a statistic summary of data
display(flavors.describe())
```

```python
# Sorting by an axis:
display(flavors.sort_index(axis=1, ascending=False))
```

```python
# Sort by values
whisky.sort_values(by='region')
```

```python
# Selection using Booleans
display(whisky[whisky.Nutty >= 3])
```

```python
# Calling methods on filtered data
display(whisky[whisky.Nutty >= 3].describe())
```

<h1>Pivot Tables</h1>
<ul><li>You can create a spreadsheet-style pivot table as a DataFrame.</li>
    <li>You can use any NumPy aggregation function (np.sum, np.max, np.min, np.mean, np.median)</li>
    <li>Normally you use pivot tables to plot the relationships between two variables.</li></ul>


```python
# Create a spreadsheet-style pivot table as a DataFrame. 
table = pd.pivot_table(whisky, values=['Body', 
           'Sweetness', 'Smoky', 'Medicinal',
           'Tobacco', 'Honey', 'Spicy', 'Winey', 
           'Nutty', 'Malty', 'Fruity', 'Floral'], 
            columns=['region'], index=[], 
                       aggfunc=max)
display(table)
```

```python
import numpy as np

# Compare Sweetness to Body
# Fill missing values using the fill_value.
# Mean Sweetness for each Body and region

# Pandas does not support the mean aggegation function directly
# YOu need to import numpy as np and use np.mean as the aggfunc
table = pd.pivot_table(whisky, 
           values=['Sweetness'], index=['Body'], 
           columns=['region'], fill_value=0, 
           aggfunc=np.mean)

display(table)
```
