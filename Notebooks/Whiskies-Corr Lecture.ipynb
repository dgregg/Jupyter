```python
import pandas as pd


# read whisky data from file
whisky = pd.read_csv("whiskies.txt")

# calculate the correlations for the data set
cor  = whisky.corr()

# relevant part of matirx in row/column 1 to 13
corr = cor.iloc[1:13, 1:13] 

display(corr)
```

```python
# Pandas does not support the colormesh/heatmap directly
# Need  to import matpltlib.pyplot as plt
import matplotlib.pyplot as plt

# Use Matplotlib.pyplot to create a pcolormesh of the correlation matirx
plt.pcolormesh(corr)

#Show the plot
plt.show()
```

```python
# Create a pcolormesh of the correlation matirx
plt.pcolormesh(corr)

# Set locations and labels of X axis / labels are column names
plt.xticks(range(12), corr.index, rotation=45)

# Set locations and labels for Y Axis Whey Y labels are row names
plt.yticks(range(12), corr.index, rotation=-15)


# Plot will be upside down use ylim to invert it
plt.ylim(12, 0)

# Add a color bar to show what colors mean 
plt.colorbar()

#Show the plot
plt.show()
```
