```python
import pandas as pd

# Create a DataFrame from the sales_data csv file
df = pd.read_csv("sales_data.csv")
```

```python
# Process Data So it is Suitable for a bar plot
# Use group by to get total sales by day
region_sales = df.groupby('Region')['Sales'].sum()

print(region_sales)
```

```python
# Create a vertical bar chart for a series
fig = region_sales.plot(kind='bar', figsize=(15,12), 
                        title="Annual Sales by Region",
                        ylabel='Annual Sales (USD)')
fig.figure.savefig("bar_all.png")
```

```python
# Aggregating Sales Using a Pandas Crosstab
tab = pd.crosstab(
    index=df['Region'],
    columns=df['Category'],
    values=df['Sales'],
    aggfunc='sum'
)

display(tab)
```

```python
# Create a vertical bar chart
fig = tab.plot(kind='bar', figsize=(15,12), 
               title="Annual Sales for Category Types by Region", ylabel='Annual Sales (USD)')
fig.figure.savefig("bar_category.png")
```

```python
# We can create horizontal bar charts by assigning the barh string value to the kind argument
fig = tab.plot(kind='barh', figsize=(15,12), 
               title="Annual Sales for Category Types by Region")

# xlabel cannot be set when creating the plot. Use set_xlabel to set after
fig.set_xlabel('Annual Sales (USD)')
fig.figure.savefig("bah.png")
```

```python
# Create a stacked vertical bar chart
# The height of the resulting bar shows the total sales for all three categories 
fig = tab.plot(kind='bar', stacked=True, figsize=(15,12), rot=30, 
               title="Annual Sales by Region with Types", ylabel='Annual Sales (USD)')
fig.figure.savefig("bar_stacked.png")
```
