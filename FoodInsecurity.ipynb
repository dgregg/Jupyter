```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("data/FoodInsecurity.csv", index_col=0)

df["Population"] = df["Population"]/1000

plt = df.plot(kind="scatter", y='FoodInsecurityPopulation', x='MedianIncome', 
              ylabel='Food Insecurity Population',  xlabel='Median Income',
              s="Population", title = "Food Insecurity in Colorado in 2019",
              figsize=(9,6))

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("data/FoodInsecurity.csv", index_col=0)

df["FoodInsecurityPercentage"] = df["FoodInsecurityPopulation"]/df["Population"]

plt = df.plot(kind="scatter", y='FoodInsecurityPercentage', x='MedianIncome', 
              title = "Food Insecurity in Colorado in 2019",
              figsize=(9,6))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("data/FoodInsecurity.csv", index_col=0)

#df = df.sort(df["Population"])
df = df.sort_values(by=['MedianIncome'])

plt = df.plot(kind="line", y='FoodInsecurityPopulation', x='MedianIncome', 
              ylabel='Food Insecurity Population',  xlabel='Median Income',
              title = "Food Insecurity in Colorado in 2019",
              figsize=(9,6))
```

```python
plt = df.plot(kind="hist", y='FoodInsecurityPopulation', bins=20, 
              ylabel='Food Insecurity Population',  xlabel='Food Insecurity Population',
              title = "Food Insecurity in Colorado in 2019",
              figsize=(9,6))
```
